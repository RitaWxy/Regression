# 2. 数据预处理

## one-hot编码

```python
# 需要onehot的特征
data_onehot = data[['sex','cp','fbs','restecg','exang','slope','thal']]
data_onehot
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder()
data_onehot = onehot.fit_transform(data_onehot)
data_onehot.toarray()
```

```python
data_onehot.shape
```

```python
# Onehot编码
# data = pd.get_dummies(data, columns=['cp','restecg','slope','thal'])
```

## 数据、标签划分

```python
x_data = data.drop(columns=['target'])
y_data = data['target']
print(x_data.shape,y_data.shape)
```

## 归一化

```python
data.std()
```

```python
x_data.shape
```

```python
stander = StandardScaler()
after_scale = stander.fit_transform(x_data.drop(columns=['sex','cp','fbs','restecg','exang','slope','thal']))
after_scale
```

```python
np.mean(after_scale,axis = 0)
```

```python
np.std(after_scale,axis = 0)
```

```python
# 合并数据
X_data = np.concatenate((after_scale,data_onehot.toarray()),axis = 1)
X_data.shape
```

## 训练集，测试集划分

```python
x_train,x_test,y_train,y_test = train_test_split(X_data,y_data,test_size = 0.2,random_state=33,shuffle = True)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

# 3. 模型定义

```python
model = LogisticRegression(random_state=0,solver='lbfgs',max_iter=1000)
model.fit(x_train,y_train)
```

```python
print(model.classes_)
print(model.coef_.shape)
```

```python
model.predict(x_test[0].reshape(-1,x_test.shape[1]))
```

```python
y_test = np.array(y_test)
```
